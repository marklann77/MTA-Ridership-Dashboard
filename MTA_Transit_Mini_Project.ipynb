```python
# We start simply by importing our csv (left hand side) into Colab
# Then, we import pandas

import pandas as pd #pd could be anything, but we're using pd as a simple way of referencing it
```

```python
# Now we need to get pandas to read our csv file & display the first five rows

df = pd.read_csv('/content/MTA_Daily_Ridership_Data__2020_-_2025_20250304.csv') # We make the 'df' variable so that the data is stored somewhere for us to call on later if we want
df.head() # First 5 rows of our csv table is displayed
```

```python
# Since we have a 'Date' column, we should check the time range of our dataset
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime format
df['Date'].min(), df['Date'].max()  # Find earliest & latest date
```

```python
# Quick review:
# What does the . (dot) between pd and to_datetime do? --> The dot means "access this function from the pandas library."

# pd → This is pandas, the library we imported earlier (import pandas as pd).
# to_datetime --> This is a function inside pandas that converts text to real date format.
# . → The dot means "get this function from pandas" (pd.to_datetime).

# Think of it like this:

# pd is like a toolbox
# to_datetime() is one of the tools inside the toolbox
# The . lets us pull the tool out to use it

import math
math.sqrt(16)  # Calls the square root function from the math library

# math = The library (like pd)
# . = "Look inside this library for..."
# sqrt() = The function (like to_datetime())
```
